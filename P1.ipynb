```python
import pandas as pd

bollywood = pd.read_csv('cleaned_bollywood_movies.csv', index_col=0)
tollywood = pd.read_csv('cleaned_tollywood_movies.csv', index_col=0)
indian = pd.read_csv('cleaned_indian_movies.csv', index_col=0)
```

```python
#calculate profit and ROI 
for df in [bollywood, tollywood, indian]:
    df['Profit'] = df['BoxOffice (Crores)'] - df['Budget (Crores)']
    df['ROI'] = df['Profit'] / df['Budget (Crores)']
```

```python
#top 3 by ROI
combined = pd.concat([bollywood, tollywood, indian])
top3_all = combined.sort_values(by='ROI', ascending=False).head(3)
print(top3_all)
```

```python
top3_hindi = bollywood.sort_values(by='ROI', ascending=False).head(3)
```

```python
top3_telugu = tollywood.sort_values(by='ROI', ascending=False).head(3)
```

```python
import matplotlib.pyplot as plt

# Plot Top 3 across all
plt.figure(figsize=(8,5))
plt.bar(top3_all['Title'], top3_all['ROI'], color='skyblue')
plt.title('Top 3 Movies by ROI (All Industries)')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Top 3 Hindi
plt.figure(figsize=(8,5))
plt.bar(top3_hindi['Title'], top3_hindi['ROI'], color='green')
plt.title('Top 3 Hindi Movies by ROI')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Top 3 Telugu
plt.figure(figsize=(8,5))
plt.bar(top3_telugu['Title'], top3_telugu['ROI'], color='orange')
plt.title('Top 3 Telugu Movies by ROI')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
